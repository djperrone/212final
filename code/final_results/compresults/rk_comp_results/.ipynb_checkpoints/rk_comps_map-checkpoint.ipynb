```python
rk_df10 = pd.read_csv("rk_comp_word10-293.csv")
```

```python
rk_df50 = pd.read_csv("rk_comp_phrase50-293.csv")
```

```python
rk_df50['pat_len'] = [50 for i in rk_df50.index]
```

```python
rk_df100 = pd.read_csv("rk_comp_phrase100-293.csv")
```

```python
rk_df100['pat_len'] = [100 for i in rk_df100.index]
```

```python
rk_df1000 = pd.read_csv("rk_comp_char1000-293.csv")
```

```python
rk_df1000['pat_len'] = [1000 for i in rk_df1000.index]
```

```python
df_list = [rk_df1,rk_df10,rk_df50,rk_df100,rk_df1000]
```

```python
text_len = []
pat_len = []
comps = []
```

```python
rk_df1000
```

```python
rows = []
cols = []
for df in df_list:
    for i in range(0,len(df.index)):  
        text_len.append(df['file_name'].values[i])
        comps.append(df['total_comps'].values[i])
        pat_len.append(df['pat_len'].values[i])

```

```python
rk_avg_df = pd.DataFrame(columns =['text_len','pat_len','total_comps'])
```

```python
rk_avg_df
```

```python
rk_avg_df['text_len']=[i for i in text_len]
rk_avg_df['pat_len']=[i for i in pat_len]
rk_avg_df['total_comps']=[i for i in comps]
```

```python
rk_avg_df
```

```python
rk_matrix = rk_avg_df.pivot('text_len','pat_len','total_comps')
```

```python
rk_avg_df['total_comps'].min()
```

```python
heatmap = sns.heatmap(data=rk_matrix,cmap='BuPu',vmin=rk_avg_df['total_comps'].min(),vmax=rk_avg_df['total_comps'].max()).invert_yaxis()
```
