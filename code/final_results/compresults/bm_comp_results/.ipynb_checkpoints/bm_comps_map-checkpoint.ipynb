```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib 
from matplotlib import pyplot as plt
```

```python
data_url = "bm_comp_word1.csv"
```

```python
bm_df1 = pd.read_csv(data_url)
```

```python
bm_df1.columns
```

```python
bm_df1['pat_len'] = [1 for i in bm_df1.index]
```

```python
bm_df10 = pd.read_csv("bm_comp_word10.csv")
```

```python
bm_df10['pat_len'] = [10 for i in bm_df10.index]
```

```python
bm_df50 = pd.read_csv("bm_comp_phrase50.csv")
```

```python
bm_df50['pat_len'] = [50 for i in bm_df50.index]
```

```python
bm_df100 = pd.read_csv("bm_comp_phrase100.csv")
```

```python
bm_df100['pat_len'] = [100 for i in bm_df100.index]
```

```python
bm_df1000 = pd.read_csv("bm_comp_char1000.csv")
```

```python
bm_df1000['pat_len'] = [1000 for i in bm_df1000.index]
```

```python
df_list = [bm_df1,bm_df10,bm_df50,bm_df100,bm_df1000]
```

```python
text_len = []
pat_len = []
comps = []
```

```python
bm_df1000
```

```python
rows = []
cols = []
for df in df_list:
    for i in range(0,len(df.index)):  
        text_len.append(df['file_name'].values[i])
        comps.append(df['total_comps'].values[i])
        pat_len.append(df['pat_len'].values[i])

```

```python
bm_avg_df = pd.DataFrame(columns =['text_len','pat_len','total_comps'])
```

```python
bm_avg_df
```

```python
bm_avg_df['text_len']=[i for i in text_len]
bm_avg_df['pat_len']=[i for i in pat_len]
bm_avg_df['total_comps']=[i for i in comps]
```

```python
bm_avg_df
```

```python
bm_matrix = bm_avg_df.pivot('text_len','pat_len','total_comps')
```

```python
bm_avg_df['total_comps'].min()
```

```python
heatmap = sns.heatmap(data=bm_matrix,cmap='BuPu',vmin=bm_avg_df['total_comps'].min(),vmax=bm_avg_df['total_comps'].max()).invert_yaxis()
```
