```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib 
from matplotlib import pyplot as plt
```

```python
data_url = "brute_comp_word1.csv"
```

```python
brute_df1 = pd.read_csv(data_url)
```

```python
brute_df1.columns
```

```python
brute_df1['pat_len'] = [1 for i in brute_df1.index]
```

```python
brute_df10 = pd.read_csv("brute_comp_word10.csv")
```

```python
brute_df10['pat_len'] = [10 for i in brute_df10.index]
```

```python
brute_df50 = pd.read_csv("brute_comp_phrase50.csv")
```

```python
brute_df50['pat_len'] = [50 for i in brute_df50.index]
```

```python
brute_df100 = pd.read_csv("brute_comp_phrase100.csv")
```

```python
brute_df100['pat_len'] = [100 for i in brute_df100.index]
```

```python
brute_df1000 = pd.read_csv("brute_comp_char1000.csv")
```

```python
brute_df1000['pat_len'] = [1000 for i in brute_df1000.index]
```

```python
df_list = [brute_df1,brute_df10,brute_df50,brute_df100,brute_df1000]
```

```python
text_len = []
pat_len = []
comps = []
```

```python
brute_df1000
```

```python
rows = []
cols = []
for df in df_list:
    for i in range(0,len(df.index)):  
        text_len.append(df['file_name'].values[i])
        comps.append(df['total_comps'].values[i])
        pat_len.append(df['pat_len'].values[i])

```

```python
brute_avg_df = pd.DataFrame(columns =['text_len','pat_len','total_comps'])
```

```python
brute_avg_df
```

```python
brute_avg_df['text_len']=[i for i in text_len]
brute_avg_df['pat_len']=[i for i in pat_len]
brute_avg_df['total_comps']=[i for i in comps]
```

```python
brute_avg_df
```

```python
brute_matrix = brute_avg_df[-5:].pivot('text_len','pat_len','total_comps')
```

```python
brute_avg_df['total_comps'].min()
```

```python
heatmap = sns.heatmap(data=brute_matrix,cmap='BuPu',vmin=brute_avg_df['total_comps'][-5:].min(),vmax=brute_avg_df[-5:]['total_comps'].max()).invert_yaxis()
```
